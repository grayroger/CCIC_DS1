```python
import pandas as pd
import seaborn as sns
import random as rd
import matplotlib.pyplot as plt
from math import sqrt

df = pd.read_csv("tbell.csv")

df.dropna(axis = 0, inplace = True, how = "any")
df.drop(columns=["Unnamed: 0", "ID","State_Purchased"], inplace=True)
df.head(2)
```

```python
for i in range(len(df['Gender'])):
    if df.iat[i,0]=='Female':
        df.iat[i,0]='1'
    else:
        df.iat[i,0]='0'
for i in range(len(df['Ever_Married'])):
    if df.iat[i,1]=='Yes':
        df.iat[i,1]='1'
    else:
        df.iat[i,1]='0'
for i in range(len(df['Graduated'])):
    if df.iat[i,3]=='Yes':
        df.iat[i,3]='1'
    else:
        df.iat[i,3]='0'
        
for i in range(len(df['Profession'])):
    if df.iat[i,4]=='Engineer':
        df.iat[i,4]='0'
    if df.iat[i,4]=='Healthcare':
        df.iat[i,4]='1'
    if df.iat[i,4]=='Executive':
        df.iat[i,4]='2'
    if df.iat[i,4]=='Doctor':
        df.iat[i,4]='3'
    if df.iat[i,4]=='Artist':
        df.iat[i,4]='4'
    if df.iat[i,4]=='Lawyer':
        df.iat[i,4]='5'
    if df.iat[i,4]=='Entertainment':
        df.iat[i,4]='6'
    if df.iat[i,4]=='Homemaker':
        df.iat[i,4]='7'
    if df.iat[i,4]=='Marketing':
        df.iat[i,4]='8'
        
for i in range(len(df)):
    if df.iat[i,6]=='Low':
        df.iat[i,6]='0'
    if df.iat[i,6]=='Average':
        df.iat[i,6]='1'
    if df.iat[i,6]=='High':
        df.iat[i,6]='2'
        

df["Gender"] = df["Gender"].astype(float)            #1 is female 0 is male
df["Ever_Married"] = df["Ever_Married"].astype(float)#1 is yes 0 is no
df["Graduated"] = df["Graduated"].astype(float)      #1 is yes 0 is n
df["Profession"] = df["Profession"].astype(float)
df["Spending_Score"] = df["Spending_Score"].astype(float)
```

```python
cor = df.corr()
plt.figure(figsize = (10,6))
sns.heatmap(cor,annot=True)
```

```python
from sklearn.cluster import KMeans
ks=[]
for i in range(1,3):
    ks.append(i)
df = df.copy([['Spending_Score', 'Age']])
```

```python
wcss = []


for i in range(1, 3):
    clustering = KMeans(n_clusters=i, init='k-means++', random_state=42)
    clustering.fit(df)
    wcss.append(clustering.inertia_)
    
data = {"KValue":ks,"Variation":wcss}
df2 = pd.DataFrame(data)
sns.lineplot(data=df2, x = "KValue", y = "Variation");
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,5))
#sns.scatterplot(ax=axes[0], data=df, x=df['Profession'], y=df['Age']).set_title('Without clustering')
sns.scatterplot(ax=axes[1], data=df, x=df['Profession'], y=df['Family_Size'], hue=clustering.labels_).set_title('Using the elbow method');
```
